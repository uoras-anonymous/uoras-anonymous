```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
outfile_dict = {
    '2subd-mlp10subd': 'results_test/test_tol_2_subdomains_mlp10meshrange-tolvalid20-exp_20260409_122028.csv',
    '2subd-mlp10subd_stat': 'results_test/test_tol_stationary_2_subdomains_mlp10stationary_20260428_092025.csv',
    '2subd-t2': 'results_test/test_tol_2_subdomains_t2_20260409_205040.csv',
    '2subd-oo2': 'results_test/test_tol_2_subdomains_oo2_20260408_122827.csv',
    '10subd-mlp10subd': 'results_test/test_tol_10_subdomains_mlp10meshrange-tolvalid20-exp_20260409_122010.csv',
    '10subd-mlp10subd_stat': 'results_test/test_tol_stationary_10_subdomains_mlp10stationary_20260425_111934.csv',
    '10subd-t2': 'results_test/test_tol_10_subdomains_t2_20260408_150315.csv',
    '10subd-oo2': 'results_test/test_tol_10_subdomains_oo2_20260408_150040.csv',
    '20subd-mlp10subd': 'results_test/test_tol_20_subdomains_mlp10meshrange-tolvalid20-exp_20260409_121954.csv',
    '20subd-mlp10subd_stat': 'results_test/test_tol_stationary_20_subdomains_mlp10stationary_20260428_092011.csv',
    '20subd-t2': 'results_test/test_tol_20_subdomains_t2_20260408_124019.csv',
    '20subd-oo2': 'results_test/test_tol_20_subdomains_oo2_20260408_123725.csv'
}
```

```python
results = pd.DataFrame()
for k, path in outfile_dict.items():
    result = pd.read_csv(path)
    result['name'] = k
    results = pd.concat([results, result], axis=0)
```

```python
results = results[results['num_dofs'] >= 78400]
```

```python
def tot_sd(x_mean, x_sd):
    var_in = np.mean(x_sd**2)
    var_between = np.mean((x_mean - np.mean(x_mean))**2)
    return np.sqrt(var_in + var_between)

results_df = results.groupby(['name']).apply(
    lambda g: pd.Series({
        'total_num_iters_mean': g['num_iters_tol_mean'].mean(),
        'total_num_iters_sd': tot_sd(g['num_iters_tol_mean'], g['num_iters_tol_sd']),
        'total_runtimes_mean': g['runtimes_mean'].mean(),
        'total_runtimes_sd': tot_sd(g['runtimes_mean'], g['runtimes_sd'])
    }),
    include_groups=False
).reset_index()
print(results_df)
```
